```python
import pandas as pd
from PIL import Image, ImageTk
import datetime
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import random

janela_principal = tk.Tk()

arquivo = 'Dados-produtos.xlsx'
arquivo1 = 'Relatorio.xlsx'
dados_produtos = pd.read_excel(arquivo)
relatorio = pd.read_excel(arquivo1)
#display(dados_produtos)

class app():
    def __init__(self):
        self.janela_principal = janela_principal
        self.main()
        self.menu()
        janela_principal.mainloop()
 
    def caixa_de_mensagem(self, sms):
        messagebox.showinfo('Aviso!!', sms)
        pass
        
    def exibir_dados_produtos(self, dataframe):
            # Limpando o Treeview
            for item in self.tree.get_children():
                self.tree.delete(item)
            
            # Adicionando os dados do DataFrame no Treeview
            for _, row in dataframe.iterrows():
                self.tree.insert("", tk.END, values=list(row))

    def main(self):
        self.vcmd = self.janela_principal.register(self.validar_entrada)
        
        self.janela_principal.title('Gerenciamento de Stock')
        self.janela_principal.configure(background='#1e3743')
        self.janela_principal.geometry('1370x730')
        self.janela_principal.resizable(False, True)
        self.janela_principal.minsize(width=1300, height=350)
        
        self.frame1 = tk.Frame(self.janela_principal)
        self.frame1.place(relx= 0.01,rely= 0.48, relwidth= 0.97, relheight= 0.5)
        self.frame1.configure(background='lightblue')
        
        self.frame2 = tk.Frame(self.janela_principal)
        self.frame2.place(relx= 0.01,rely= 0.03, relwidth= 0.68, relheight= 0.4)
        
        self.img = Image.open('foto2.jpg')
        self.img = self.img.resize((600, 270))
        self.img_tk = ImageTk.PhotoImage(self.img)
        self.fundo = tk.Label(self.frame2, image=self.img_tk)
        self.fundo.pack()
         
        self.frame3 = tk.Frame(self.janela_principal)
        self.frame3.place(relx= 0.7,rely= 0.03, relwidth= 0.28, relheight= 0.4)
        
        self.tree = ttk.Treeview(self.frame1, columns=list(dados_produtos.columns), show="headings")
        self.tree.place(relx=0.01, rely=0.04, relwidth=0.962, relheight=0.85)
        for coluna in dados_produtos.columns:
            self.tree.heading(coluna, text=coluna)
            self.tree.column(coluna, anchor=tk.CENTER)
            
        rol = tk.Scrollbar(self.frame1, orient='vertical', command=self.tree.yview)
        self.tree.configure(yscrollcommand=rol.set)
        rol.place(relx=0.972, rely=0.04, relwidth=0.015, relheight=0.85)
        
        rol_horiz = tk.Scrollbar(self.frame1, orient='horizontal', command=self.tree.xview)
        self.tree.configure(xscrollcommand=rol_horiz.set)
        rol_horiz.place(relx=0.01, rely=0.83, relwidth=0.962, relheight=0.06)
            
        self.btn_confirmar0 = tk.Button(self.frame3, text='  Confirmar  ', background='lightblue', bd=5, command=self.vender)
        self.btn_confirmar0.place(relx= 0.42, rely= 0.3)
        
        self.btn_exibir = tk.Button(self.frame1, text="Exibir Dados dos Produtos", command=lambda: self.exibir_dados_produtos(dados_produtos), activebackground='lightblue')
        self.btn_exibir.place(relx=0.43, relwidth=0.15, rely=0.9, relheight=0.08)
        
        self.btn_confirmar = tk.Button(self.frame3, text='  Confirmar  ', background='lightblue', bd=5, command=self.confirmar_venda)
        self.btn_confirmar.place(relx= 0.7, rely= 0.87)
        
        self.btn_confirmar1 = tk.Button(self.frame2, text='     Confirmar     ', background='lightblue', bd=5, command=self.confirmar_adicionar_produto)
        
        self.btn_cancelar = tk.Button(self.frame3, text='    Cancelar    ', activebackground='lightblue', bd=5, command=self.func_cancelar)
        self.btn_cancelar.place(relx= 0.45, rely= 0.87)
        
        self.lb_codigo = tk.Label(self.frame3, text='   Vender   ', background='lightblue')
        self.lb_codigo.place(rely=0, relwidth=1)
        self.lb_buscar = tk.Label(self.frame3, text='  Buscar por Cod.                     Buscar por Nome')
        self.lb_buscar.place(rely=0.08, relx=0.01)
        
        self.ent_codigo = tk.Entry(self.frame3, validate='key', validatecommand=(self.vcmd, "%P"))
        self.ent_codigo.place(rely=0.16, relx=0.03, relwidth=0.25)
        
        self.ent_nome = tk.Entry(self.frame3)
        self.ent_nome.place(rely=0.16, relx=0.42, relwidth=0.35)
        
        self.lb_qtd = tk.Label(self.frame3, text='  Quantidade')
        self.lb_qtd.place(rely=0.26, relx=0.01)
        
        self.ent_qtd = tk.Entry(self.frame3, validate='key', validatecommand=(self.vcmd, "%P"))
        self.ent_qtd.place(rely=0.33, relx=0.03, relwidth=0.2)
        
        self.linha = tk.Label(self.frame3, background='lightblue')
        self.linha.place(rely=0.44, relwidth=1, relheight=0.01)

    def validar_entrada(self, texto):
        return texto.isdigit() or texto == ""
    
    def func_cancelar(self):
        self.ent_codigo.delete(0, tk.END)
        self.ent_nome.delete(0, tk.END)
        self.ent_qtd.delete(0, tk.END)
        self.dados.destroy()
        
    def vender(self):
        self.cod = self.ent_codigo.get()
        self.nome = self.ent_nome.get()
        self.qtd = self.ent_qtd.get()

        if self.qtd == "":
            self.qtd = 1
        else:
            self.qtd = int(self.qtd)
            
        try:
            if self.cod:
                self.produto = dados_produtos[dados_produtos["Código"] == int(self.cod)]
            else:
                # Filtrar produto pelo nome, se código não for fornecido
                self.produto = dados_produtos[dados_produtos["Nome"].str.lower() == self.nome.lower()]
                self.cod = self.produto.iloc[0]["Código"]
                
            if not self.produto.empty:
                self.nome = self.produto.iloc[0]["Nome"]
                preco = self.produto.iloc[0]["Preço Unitário"]
                total = preco * self.qtd

                # Exibir detalhes da venda
                if hasattr(self, 'dados'):
                    self.dados.destroy()  # Remove o label anterior se existir
                    
                self.dados = tk.Label(
                    self.frame3, 
                    text=f'Produto: {self.nome} \nCódigo: {self.cod} \nPreço: {preco:.2f} kzs \nQuantidade: {self.qtd} \nTotal: {total:.2f} kzs', 
                    font=('verdana', 12, 'bold'), 
                    justify='left',
                    background='lightblue'
                )
                self.dados.place(rely=0.48, relx=0.05)

            else:
                msg = 'Produto não encontrado!'
                self.caixa_de_mensagem(msg)
        except IndexError:
            msg = 'Produto não encontrado!'
            self.caixa_de_mensagem(msg)

    def confirmar_venda(self):
        for i in dados_produtos.itertuples():
            if i.Nome == self.nome:
                 linha = i.Index
                 break
                
        if self.qtd > self.produto.iloc[0]["Quantidade"]:
            if hasattr(self, 'dados'):
                    self.dados.destroy()
            self.dados = tk.Label(
                    self.frame3, 
                    text=f'Sem Produto Suficiente \nPara Sastifazer o Pedido!\n\nQuantidade Disponível: {self.produto.iloc[0]["Quantidade"]}', 
                    font=('verdana', 12, 'bold'), 
                    justify='left',
                    background='lightblue')
            self.dados.place(rely=0.48, relx=0.05)
        else:
            dados_produtos.at[linha, 'Quantidade'] = int(self.produto.iloc[0]["Quantidade"]) - self.qtd
            if hasattr(self, 'dados'):
                    self.dados.destroy()
            self.dados = tk.Label(
                    self.frame3, 
                    text=f'Vendido!', 
                    font=('verdana', 13, 'bold'), 
                    justify='left',
                    background='lightblue')
            self.dados.place(rely=0.48, relx=0.05)
            dados_produtos.to_excel(arquivo, index=False)
            
            #Criar o Relatorio
            data = datetime.datetime.now().strftime('%d/%m/%Y %H:%M:%S')
            relatorio.loc[relatorio.shape[0]] = [self.cod, self.nome, self.produto.iloc[0]["Categoria"], self.qtd, data]
            relatorio.to_excel(arquivo1, index=False)
            #display(relatorio)
    
    def menu(self):
        menus_ =  tk.Menu(janela_principal, background='blue', font=('verdana', 12, 'bold'))
        janela_principal.config(menu=menus_)
        
        def Sair(): janela_principal.destroy() 
        
        menu1 = tk.Menu(menus_)
        menus_.add_cascade(label='Opções', menu=menu1)
        menu1.add_command(label='Novo Produto', command=self.novo_produto)
        menu1.add_command(label='Entrada de Produtos', command=self.entrada_produto)
        menu1.add_command(label='Remover Produto', command=self.remover_produto)
        menu1.add_command(label='Sair', command=Sair)
        
    def entrada_produto(self):
        self.main()
        dados_produtos.to_excel(arquivo, index=False)
        
    def novo_produto(self):
        self.img = self.img.resize((400, 210))
        self.img_tk = ImageTk.PhotoImage(self.img)
        self.fundo = tk.Label(self.frame2, image=self.img_tk)
        self.fundo.place(relx=0.43, rely=0.2)
        
        self.sms = tk.Label(
                    self.frame2, 
                    text=f'Adicionar Novo Produto', 
                    font=('verdana', 13, 'bold'), 
                    justify='left',
                    background='lightblue')
        self.sms.place(rely=0.05, relx=0.03, relwidth=0.93)
        
        self.texto = tk.Label(
                    self.frame2, 
                    text=f'Nome \n\n\n\nCódigo                                 Categoria \n\n\n\nQuantidade Inicial                                Preço Unitário ', 
                    justify='left')
        self.texto.place(relx=0.03, rely=0.2)
        
        self.ent_nome1 = tk.Entry(self.frame2)
        self.ent_nome1.place(relx=0.03, rely=0.3, relwidth=0.35)
        
        self.ent_cod1 = tk.Entry(self.frame2, validate='key', validatecommand=(self.vcmd, "%P"))
        self.ent_cod1.place(relx=0.03, rely=0.5, relwidth=0.1)
        
        self.ent_cat1 = tk.Entry(self.frame2)
        self.ent_cat1.place(relx=0.18, rely=0.5, relwidth=0.15)
        
        self.ent_qtd1 = tk.Entry(self.frame2, validate='key', validatecommand=(self.vcmd, "%P"))
        self.ent_qtd1.place(relx=0.03, rely=0.7, relwidth=0.12)
        
        self.ent_preco1 = tk.Entry(self.frame2, validate='key', validatecommand=(self.vcmd, "%P"))
        self.ent_preco1.place(relx=0.24, rely=0.7, relwidth=0.07)
        
        self.btn_confirmar1.place(relx= 0.86, rely= 0.85)
        
        dados_produtos.to_excel(arquivo, index=False)
        
    def confirmar_adicionar_produto(self):
        find = False
        if self.ent_nome1.get().strip():
            
            for item in dados_produtos['Nome']:
                if str(item).lower() == str(self.ent_nome1.get()).lower():
                    find = True
                    msg = 'O Produto já existe!!'
                    self.caixa_de_mensagem(msg)
                    break
                
            if find == False:
                if self.ent_qtd1.get().strip():
                
                    if self.ent_preco1.get().strip():
                        
                        if self.ent_cat1.get().strip():
                                
                            if not self.ent_cod1.get().strip():
                                p = False
                                while p == False:
                                    cod = str(random.randint(11111, 99999))
                                    if cod not in dados_produtos['Código']:
                                        p = True
                                
                                self.ent_cod1.delete(0, tk.END)
                                self.ent_cod1.insert(0, str(cod))
                                
                                data = datetime.datetime.now().strftime('%d/%m/%Y %H:%M:%S')
                                dados_produtos.loc[dados_produtos.shape[0]] = [self.ent_cod1.get(), 
                                                                            self.ent_nome1.get(), 
                                                                            self.ent_cat1.get(), 
                                                                            self.ent_qtd1.get(), 
                                                                            self.ent_preco1.get(), data, 0]
                                dados_produtos.to_excel(arquivo, index=False)
                                msg = 'Produto Adicionado com Sucesso😊!!'
                                self.caixa_de_mensagem(msg)
                            
                            else:
                                data = datetime.datetime.now().strftime('%d/%m/%Y %H:%M:%S')
                                dados_produtos.loc[dados_produtos.shape[0]] = [self.ent_cod1.get(), 
                                                                            self.ent_nome1.get(), 
                                                                            self.ent_cat1.get(), 
                                                                            self.ent_qtd1.get(), 
                                                                            self.ent_preco1.get(), data, 0]
                                dados_produtos.to_excel(arquivo, index=False)
                                msg = 'Produto Adicionado com Sucesso😊!!'
                                self.caixa_de_mensagem(msg)
                        else:
                            msg = 'É necessário uma categoria para o produto!!'
                            self.caixa_de_mensagem(msg)
                    else:
                        msg = 'É necessário um preço para o produto!!'
                        self.caixa_de_mensagem(msg)
                else:  
                    msg = 'É necessário uma quantidade \nincial para o produto!!'
                    self.caixa_de_mensagem(msg)
        else:
            msg = 'É necessário um nome para o produto!!'
            self.caixa_de_mensagem(msg)
                        
            
    def remover_produto():
        print
        dados_produtos.to_excel(arquivo, index=False)

    def Notificacoes():
        print
        
app()
```

```python

# Função para exibir o DataFrame no Treeview
def exibir_dataframe(dataframe):
    # Limpando o Treeview
    for item in tree.get_children():
        tree.delete(item)
    
    # Adicionando os dados do DataFrame no Treeview
    for _, row in dataframe.iterrows():
        tree.insert("", tk.END, values=list(row))

# Criando a janela principal
janela = tk.Tk()
janela.title("Exibição do DataFrame com Tkinter")
janela.geometry("600x400")
janela.configure(bg="black")

# Criando o Treeview para exibir os dados
tree = ttk.Treeview(janela, columns=list(dados_produtos.columns), show="headings")
tree.pack(expand=True, fill=tk.BOTH)

# Configurando os cabeçalhos das colunas
for coluna in dados_produtos.columns:
    tree.heading(coluna, text=coluna)
    tree.column(coluna, anchor=tk.CENTER)

# Botão para exibir o DataFrame
btn_exibir = tk.Button(janela, text="Exibir DataFrame", command=lambda: exibir_dataframe(dados_produtos))
btn_exibir.pack(pady=10)

# Executando a interface gráfica
janela.mainloop()
```
